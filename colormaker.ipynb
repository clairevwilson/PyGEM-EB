```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
import pickle
```

```python
# Define the colors for your colormap
colors = [['#cc0066','#ffffff','#ffcc00'],
          ['#000099','#ffffff','#33cc33'],
            ['#4f00a2','#ffffff','#00cc99'],
            ['#ff3300','#ffffff','#0066ff'],]
fig,axes = plt.subplots(2,2)
axes = axes.flatten()
for a,ax in enumerate(axes):
    # Create the colormap
    custom_cmap = LinearSegmentedColormap.from_list('custom_cmap', colors[a])

    # Generate data for visualization
    data = np.random.rand(10, 10)

    # Plot using the custom colormap
    cax = ax.imshow(data, cmap=custom_cmap)
    fig.colorbar(cax, ax=ax)

    # Save the colormap
    with open(f'data/cmaps/custom_divergent_cmap_{a}.pkl', 'wb') as f:
          pickle.dump(custom_cmap, f)
plt.show()
```
